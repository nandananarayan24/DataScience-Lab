```python
import matplotlib.pyplot as plt
import numpy as np

x=np.array([1,2,3])
y=np.array([4,5,6])


plt.plot(x,marker='^',color='g')
plt.plot(y,'o:b')
plt.xlabel("time")
plt.ylabel("distance")
plt.title("time-distance graph")
plt.grid(axis='y')
plt.grid(color='y',linewidth=1.5)
plt.show()
```

SUBPLOT

```python

plt.subplot(2,2,1) # 1->row divide , 2->col divide ,3->which portion
plt.plot(x)
plt.subplot(2,2,2)
plt.plot(y)
plt.subplot(2,2,3)
plt.plot(x)
plt.subplot(2,2,4)
plt.plot(y)


plt.show()
```

BMI _ AGE graph

```python
import seaborn as sns

bmi=np.array([17,18,19,20])
age=np.array([18,24,28,30])

plt.xlabel("BMI")
plt.ylabel("AGE")
plt.title("BMI-AGE graph")

plt.scatter(bmi,age)
```

SEABORN

```python
import seaborn as sns

bmi=np.array([17,18,19,20])
age=np.array([18,24,28,30])
sugar=np.array(['yes','no','yes','no'])


plt.xlabel("BMI")
plt.ylabel("AGE")
plt.title("BMI-AGE graph")

sns.scatterplot(x=bmi,y=age,hue=sugar,palette={'yes':'green','no':'red'})
```
